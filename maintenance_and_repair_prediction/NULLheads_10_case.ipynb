## Объявление функций

```python
def text_cleaning(text):
    """Функция для очистки текста от знаков и чисел"""

    # Создаем таблицу перевода для удаления символов пунктуации и цифр
    punctuation_and_digits = string.punctuation + string.digits
    translator = str.maketrans('', '', punctuation_and_digits)

    # Удаляем символы пунктуации и цифры из текста
    cleaned_text = text.translate(translator)

    return cleaned_text


def process_text(text):
    """Функция для лемматизации слов в предложениях"""

    # Инициализация стеммера и морфологического анализатора
    stemmer = RussianStemmer()
    morph_analyzer = MorphAnalyzer()

    # Токенизация текста на слова
    words = nltk.word_tokenize(text)

    # Лемматизация каждого слова
    lemmatized_words = [morph_analyzer.parse(word)[0].normal_form for word in words]

    return lemmatized_words


def filter_array(arr):
    """Функция для удаления коротких слов (менее двух букв) и оставшихся символов"""

    return [s for s in arr if len(s) > 2]


def get_coordinates(address):
    """Функция для получения координат объекта по адресу"""
    geolocator = Nominatim(user_agent="my_app")
    location = geolocator.geocode(address)
    if location:
        time.sleep(3)
        return location.latitude, location.longitude
    else:
        return None, None
    

def print_all_metrics(y_true, y_pred, proba, title='Metrics'):
    """Функция для отображения метрик модели машинного обучения"""

    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred, average='weighted', zero_division=0)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred, average='weighted', zero_division=0)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred, average='weighted', zero_division=0)))


def train_and_print_metrics(model, model_name):
    """Функция, для автоматического обучения моделей"""

    # Обучаем модель на данных
    model.fit(X_train_st, y_train)

    # Делаем предсказание
    predictions = model.predict(X_test_st)
    predictions_proba = model.predict_proba(X_test_st)

    # Печатаем метрики
    print_all_metrics(
        y_test,
        predictions,
        predictions_proba,
        title=f"Метрики для модели {model_name}:"
    )
```

## Создание словаря "проблема - решение"

**Примечание:**

В основе создания словаря лежит лемматизация текста - процесс преобразования слова в его базовую форму (лемму).

*Пример: предложение "Мама мыла раму" после процесса лемматизации будет выглядеть как "мать мыть рама".*

```python
# Очищаем текст от символов и знаков
df_common_property_work_kinds['name_lemm'] = df_common_property_work_kinds['NAME'].apply(text_cleaning)

# Лемматизируем текст
df_common_property_work_kinds['name_lemm'] = df_common_property_work_kinds['name_lemm'].apply(process_text)

# Лемматизируем текст
df_common_property_work_kinds['name_lemm'] = df_common_property_work_kinds['name_lemm'].apply(filter_array)

# Оставим только необходимые столбцы
df_common_property_work_kinds = df_common_property_work_kinds[['ID', 'NAME', 'name_lemm']]

# Выведем первые 5 строк
df_common_property_work_kinds.head()
```

```python
# Формируем массив уникальных проблем, зарегистрированных в инцедентов
problems_dict = pd.DataFrame(df_incidents.Наименование.unique()).rename(columns={0: 'problem_title'}) \
                                                                .dropna().reset_index().drop(columns='index')

# Очищаем текст от символов и знаков
problems_dict['problem_lemm'] = problems_dict.problem_title.apply(text_cleaning) 

# Лемматизируем текст
problems_dict['problem_lemm'] = problems_dict.problem_title.apply(process_text)

# Лемматизируем текст
problems_dict['problem_lemm'] = problems_dict['problem_lemm'].apply(filter_array)

# Выведем первые 5 строк
problems_dict.head(5)
```

```python
# Созадём в датафрейме новый, пустой столбец
problems_dict['solve_service_common_property'] = ''

# Перебираем все строки столбца с лемматизированными проблемами
for ind, arr in enumerate(problems_dict['problem_lemm']):

    # Объявляем массив, в который будем записывать подходящие виды работ
    solve_service = []

    # Сравниваем проблему со всеми возможными решениями проблемы. Находим множество одинаковых значений
    for idx_k, serv in enumerate(df_common_property_work_kinds['name_lemm']):
        mutual_val_len = len(set(arr) & set(serv))

        # Если количество совпадений по словам больше 1, то записываем решение, как подходящее под проблему
        if mutual_val_len >= 1:
            solve_service.append([df_common_property_work_kinds['NAME'].loc[idx_k]])

    # Записываем в строку с проблемой массив подходящих решений
    problems_dict['solve_service_common_property'].loc[ind] = solve_service
```

```python
# Выводим получившийся словарь
problems_dict.sort_values(by='solve_service_common_property', ascending=False).head(10)
```

**Примечание:**

Метод способствует лишь в некоторой мере облегчить и автоматизаровать создание словаря. Для более качеественного результата метод требует некоторой ручной экспертной обработки.

### Формирование датасета для обучения

Сделаем сводную таблицу. Индексами таблицы будут id объектов. Столбцами - возможные проблемы.
Значениями - количество зарегистрированных проблем. Если проблем на объекте не было зарегистрированно - записываем ноль.

```python
pivot_data = pd.pivot_table(df_incidents, index='unom', columns='Наименование', 
                                        values='Округ', aggfunc='count', fill_value=0) \
                                                                        .reset_index() \
                                                                        .rename(columns={'index': 'unom'})
pivot_data.head()
```

```python
# Формируем таблицу id объекта - количество проблем
problems_cnt = df_incidents.groupby('unom').agg({'Наименование': 'count'}) \
                                           .reset_index().rename(columns={'Наименование': 'problems_count'})

# Получаем характеристики объектов из таблицы со всеми объектами
pred_unoms = pivot_data.unom
predict_data = df_specifications.query('unom in @pred_unoms')

# Выделяем столбцы со строковым типом данных и удаляем их
object_columns = predict_data.select_dtypes(include=['object']).columns
predict_data = predict_data.drop(object_columns, axis=1)

# Объедяняем со сводной таблицей
predict_data = predict_data.merge(pivot_data, on='unom')

# Добавляем столбец с общим количеством проблем на объекте
predict_data = predict_data.merge(problems_cnt, on='unom')

# Удалим ненужные столбцы и пустые значения
predict_data = predict_data.drop(columns=to_delete_data)
predict_data = predict_data.dropna(axis=1, how='all')
predict_data = predict_data.dropna(axis=0)

# Выведем первые 5 строк
predict_data.head()
```
